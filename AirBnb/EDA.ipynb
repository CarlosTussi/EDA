# Removing Columns with more than 65% of missing values

```python
percentage = 65
df_processed = df.drop(df.isna().sum()[df.isna().sum() > (len(df)*percentage)/100].index, axis = 1)
```

```python
df_processed.isna().sum()[df_processed.isna().sum() > 0].sort_values()
```

# Removing columns

#### Checking unique values of columns

```python
unique_values_sum = [(len(value), item) for item, value in df_processed.apply(lambda x : pd.unique(x), axis = 0).items()]
unique_values_sum.sort(key=lambda x: x[0])

unique_values_sum
```

#### Removing columns with same value for all the rows

```python
columns_same_value = [elem[1] for elem in unique_values_sum if elem[0]==1]
columns_same_value
```

```python
df_processed = df_processed.drop(columns_same_value, axis = 1)
```

#### Removing reduntant information

```python
#Identifying reduntant columns

print("Equal values: " + str((df.host_listings_count == df.host_total_listings_count).sum()))
print("Diferent values: ") #The different values are actuawlly the missing values
df.loc[(df.host_listings_count != df.host_total_listings_count),["host_listings_count","host_total_listings_count"]]
```

```python
#Dropping reduntant columns

df_processed = df_processed.drop("host_total_listings_count", axis = 1)
```

```python
unique_values_sum = [(len(value), item) for item, value in df_processed.apply(lambda x : pd.unique(x), axis = 0).items()]
unique_values_sum.sort(key=lambda x: x[0])

unique_values_sum
```

#### Removing URLs

```python
#Look for columns that end with url
pattern_url = re.compile(".+_url")
url_features = [col for col in df.columns if re.search(pattern_url, col)]

#Drop columns that contain url
df_processed = df_processed.drop(url_features, axis = 1)
```

#### Removing unique identifiers

```python
#Dropping unique identifiers (not usefull for the model)

pattern_id = re.compile(".*id.*")

unique_identifiers = [col for col in df_processed.columns if re.search(pattern_id, col)]
unique_identifiers
```
