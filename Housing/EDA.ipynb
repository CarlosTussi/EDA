3) House value for each category of houses based on the approximity of the ocean

```python
grp_ocean_proximity = df.groupby('ocean_proximity')         #GroupBy category
result_mean = grp_ocean_proximity["median_house_value"].mean()   #Get the mean groupping by categroy
result_std = grp_ocean_proximity["median_house_value"].std()   #Adding another colums with the sum aggregation
rdf = pd.DataFrame({"mean house_value": result_mean.values,
                    "std": result_std.values}, index = result_std.index)
rdf

```

8. Pie chart with the proportion of each ocean_proximity

```python
#Create an array with the count of each ocean proximity
count_ocean_proximity = df.groupby("ocean_proximity")["ocean_proximity"].count()

pie_values = count_ocean_proximity.values
pie_labels = count_ocean_proximity.index

def absolute_value(percentage_slice): 
    total = sum(pie_values) #100% of the total
    absolute = int(percentage_slice/100.*total)
    return f'{absolute}'

plt.pie(pie_values, labels = pie_labels, autopct = absolute_value)
```

9) Bar chart average price according to proximity

```python
proximity = df.ocean_proximity.unique()
price = df.groupby("ocean_proximity")["median_house_value"].mean()

x = proximity
y = price.values

plt.bar(x, y)
```
